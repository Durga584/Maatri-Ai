# Maternal Health Risk Prediction - Exploratory Data Analysis (EDA)

This notebook performs EDA on the Maternal Health Risk Data Set to understand distributions, correlations, outliers, and class imbalance.

### Dataset Features:
1. **Age**: Age in years when pregnant.
2. **SystolicBP**: Upper value of Blood Pressure in mmHg.
3. **DiastolicBP**: Lower value of Blood Pressure in mmHg.
4. **BS**: Blood glucose levels (Blood Sugar) in mmol/L.
5. **BodyTemp**: Body temperature in degrees Fahrenheit.
6. **HeartRate**: Resting heart rate in beats per minute.
7. **RiskLevel**: Target variable indicating low, mid, or high pregnancy risk.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure inline plotting
%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Load and Inspect Dataset

```python
csv_path = os.path.join("..", "datasets", "maternal_health_risk.csv")

# If dataset doesn't exist yet, download it
if not os.path.exists(csv_path):
    from ml.risk_prediction.preprocessing import clean_and_inspect_data
    import urllib.request
    url = "https://raw.githubusercontent.com/Ugochukwuodinaka/Maternal-Health-Risk-Analysis/main/Maternal%20Health%20Risk%20Data%20Set.csv"
    os.makedirs(os.path.dirname(csv_path), exist_ok=True)
    urllib.request.urlretrieve(url, csv_path)
    print("Downloaded dataset.")

df = pd.read_csv(csv_path)
df.head()
```

### Basic Properties, Types, and Missing Values

```python
print("=== Dataset Shape ===")
print(df.shape)
print("\n=== Info ===")
print(df.info())
print("\n=== Missing Values ===")
print(df.isnull().sum())
print("\n=== Duplicate Count ===")
print(df.duplicated().sum())
```

## 2. Statistical Analysis

```python
print("=== Descriptive Statistics ===")
df.describe().T
```

## 3. Target Variable Analysis (Class Imbalance)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='RiskLevel', data=df, order=['low risk', 'mid risk', 'high risk'])
plt.title('Target variable: RiskLevel Distribution')
plt.show()

print("Percentage distribution:")
df['RiskLevel'].value_counts(normalize=True) * 100
```

## 4. Feature Distributions & Outliers

```python
features = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']

# Histograms
df[features].hist(bins=15, figsize=(15, 10))
plt.suptitle('Feature Distributions')
plt.show()

# Boxplots for outlier detection
plt.figure(figsize=(15, 6))
for i, col in enumerate(features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Outliers in {col}')
plt.tight_layout()
plt.show()
```

## 5. Correlation Analysis

```python
# Convert RiskLevel to categorical codes for correlation matrix
df_corr = df.copy()
risk_map = {'low risk': 0, 'mid risk': 1, 'high risk': 2}
df_corr['RiskLevel_encoded'] = df_corr['RiskLevel'].str.strip().str.lower().map(risk_map)

plt.figure(figsize=(8, 6))
sns.heatmap(df_corr.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
